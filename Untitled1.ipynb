```python
import pandas as pd

# Load behavioral data (as per user's path)
behav = pd.read_csv("/home/jovyan/cache/memoryNSD/sub01_behav.tsv", sep='\t')

# Filter for shared1000 images only
shared = behav[behav["is_shared1000"] == 1]

# Separate encoding and retrieval trials
encoding = shared[shared["shared1000_repNum"] == 1].copy()
retrieval1 = shared[shared["shared1000_repNum"] == 2].copy()

# Match encoding and retrieval trials by image ID (10KID)
item_pairs = pd.merge(
    encoding,
    retrieval1,
    on="10KID",
    suffixes=("_enc", "_ret")
)

# Show results
item_pairs[["10KID", "TRIAL_enc", "TRIAL_ret", "ISCORRECT_enc", "ISCORRECT_ret"]].head()
```
